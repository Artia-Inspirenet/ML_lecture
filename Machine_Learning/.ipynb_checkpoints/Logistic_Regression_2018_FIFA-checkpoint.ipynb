```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# 경기 결과 DataFrame
results = pd.read_csv('/home/dongsu/FIFA-2018-World-cup-predictions/datasets/results.csv')
```

```python
results.head()
```

## 이긴 팀 column 추가

```python
winner = []
for i in results.itertuples():
    if i.home_score > i.away_score:
        winner.append(i.home_team)
    elif i.home_score < i.away_score:
        winner.append(i.away_team)
    else:
        winner.append('Draw')
results['winning_team'] = winner
```

```python
results.head()
```

## 한국팀의 경기 성적은?

```python
df = results[(results['home_team'] == 'Korea Republic') | (results['away_team'] == 'Korea Republic')]
korea = df.iloc[:]
korea.head()
```

```python
korea.shape
```

```python
win = (korea['winning_team'] == 'Korea Republic').sum()
draw = (korea['winning_team'] == 'Draw').sum()
loss = korea.shape[0] - (win + draw)
```

```python
label = ('Win', 'Draw', 'Lose')
y_pos = np.arange(len(label))
result = [win, draw, loss]

plt.bar(y_pos, result, align='center', alpha=0.5)
plt.xticks(y_pos, label)
plt.title('Game Result') 
plt.show()
```

## 국가  축소

```python
total_home = results.home_team.unique()
```

```python
total_away = results.away_team.unique()
```

```python
# 총 245 개국
len(set(np.concatenate([total_home, total_away])))
```

```python
# 선택 : 32 개국
worldcup_teams = ['Australia', ' Iran', 'Japan', 'Korea Republic', 
            'Saudi Arabia', 'Egypt', 'Morocco', 'Nigeria', 
            'Senegal', 'Tunisia', 'Costa Rica', 'Mexico', 
            'Panama', 'Argentina', 'Brazil', 'Colombia', 
            'Peru', 'Uruguay', 'Belgium', 'Croatia', 
            'Denmark', 'England', 'France', 'Germany', 
            'Iceland', 'Poland', 'Portugal', 'Russia', 
            'Serbia', 'Spain', 'Sweden', 'Switzerland']
```

```python
df_32 = results[results['home_team'].isin(worldcup_teams) | results['away_team'].isin(worldcup_teams)]
df_32.head()
```

```python
df_32.shape
```

## 1930년 이후의 경기만 고려

```python
def get_year(x):
    return int(x[0])

year_series = df_32.date.str.split(pat='-').apply(get_year)
year_series
```

```python
# 경기 년도 column 추가
df_32['match_year'] = year_series
df_32.head()
```

```python
new_df_32 = df_32[df_32['match_year'] >= 1930].reset_index(drop=True)
new_df_32.head()
```

```python
new_df_32.shape
```

## Home 팀 VS Away 팀

```python
game_df = new_df_32.drop(['date', 'home_score', 'away_score', 'tournament', 'city', 'country', 'match_year'], axis=1)
game_df.head()
```

```python
# home_team win = 2
# draw = 1
# away_team win = 0

game_df.loc[game_df.winning_team == game_df.home_team, 'winning_team'] = 2
game_df.loc[game_df.winning_team == 'Draw', 'winning_team'] = 1
game_df.loc[game_df.winning_team == game_df.away_team, 'winning_team'] = 0

game_df.head()
```

## 현재까지의 월드컵 우승 횟수

```python
cups = pd.read_csv('/home/dongsu/FIFA-2018-World-cup-predictions/datasets/WorldCups.csv')
cups
```

```python
cups.loc[cups.Winner == 'Germany FR', 'Winner'] = 'Germany'
cups
```

```python
championships = cups.Winner.value_counts()
championships
```

```python
game_df.insert(2, 'home_champion', game_df['home_team'].map(championships).fillna(0).astype('int'))
game_df.insert(3, 'away_champion', game_df['away_team'].map(championships).fillna(0).astype('int'))
game_df.head()
```

## FIFA 랭킹

```python
ranking = pd.read_csv('/home/dongsu/FIFA-2018-World-cup-predictions/datasets/fifa_rankings.csv')
ranking.head()
```

```python
game_df.insert(4, 'home_rank', game_df['home_team'].map(ranking.set_index('Team')['Position']).fillna(212).astype('int'))
game_df.insert(5, 'away_rank', game_df['away_team'].map(ranking.set_index('Team')['Position']).fillna(212).astype('int'))
game_df.head()
```

```python
final = pd.get_dummies(game_df, prefix=['home_team', 'away_team'], columns=['home_team', 'away_team'])
final.head()
```

```python
data = final.drop(['winning_team'], axis=1)
label = final['winning_team']
label = label.astype('int')
```

```python
data_train, data_test, label_train, label_test = train_test_split(data, label, test_size=0.3, random_state=42)
```

```python
logreg = LogisticRegression()
logreg.fit(data_train, label_train)
print ('Training_set_accuracy :', '%.3f'%(logreg.score(data_train, label_train)))
```

```python
print ('Training_set_accuracy :', '%.3f'%(logreg.score(data_test, label_test)))
```

```python
pred_df = fixtures[['Home Team', 'Away Team']]
pred_df.head()
```

```python
pred_df.insert(0, 'home_champion', pred_df['Home Team'].map(championships).fillna(0).astype('int'))
pred_df.insert(1, 'away_champion', pred_df['Away Team'].map(championships).fillna(0).astype('int'))
pred_df.head()
```

```python
pred_df.insert(2, 'home_rank', pred_df['Home Team'].map(ranking.set_index('Team')['Position']).fillna(212).astype('int'))
pred_df.insert(3, 'away_rank', pred_df['Away Team'].map(ranking.set_index('Team')['Position']).fillna(212).astype('int'))
pred_df.head()
```

```python
pred_set = pd.get_dummies(pred_df, prefix=['home_team', 'away_team'], columns=['Home Team', 'Away Team'])
pred_set.head()
```

```python
# 없는 column 추가
# 값은 0으로 설정
missing_cols = set(data.columns) - set(pred_set.columns)
for c in missing_cols:
    pred_set[c] = 0
```

```python
# column 순서 정렬
pred_set = pred_set[data.columns]
```

```python
predictions = logreg.predict(pred_set)
predictions
```

```python
for i in range(fixtures.shape[0]):
    print (pred_df.loc[i, 'Home Team'] + " VS " + pred_df.loc[i, 'Away Team'])
    if predictions[i] == 2:
        print("Winner: " + pred_df.loc[i, 'Home Team'])
    elif predictions[i] == 1:
        print("Draw")
    else:
        print("Winner: " + pred_df.loc[i, 'Away Team'])
    print('Probability of ' + pred_df.loc[i, 'Home Team'] + ' winning: ', '%.3f'%(logreg.predict_proba(pred_set)[i][2]))
    print('Probability of Draw: ', '%.3f'%(logreg.predict_proba(pred_set)[i][1]))
    print('Probability of ' + pred_df.loc[i, 'Away Team'] + ' winning: ', '%.3f'%(logreg.predict_proba(pred_set)[i][0]))
    print("\n")
```
