## 1) Regression

### 1-1) Linear Regression from scratch

```python
import numpy as np
from matplotlib import pyplot as plt
% matplotlib inline
```

# Step 1. Data Creation

```python
m = 100 # 데이터 개수(sample, example)
```

```python
x = np.random.randn(1,m)
y = 3 * x + np.random.rand(1,m) * 5
```

```python
plt.scatter(x,y)
plt.show()
```

```python
print (x.shape)
print (y.shape)
```

# Step 2. Model definition

```python
def hypothesis(x, w, b):
    pred = np.dot(x, w) + b # hypothesis = W1X1 + W2X2 + ... + WnXn + b
    return pred
```

```python
def cost(x, w, b, y):
    loss = 1/2*(hypothesis(x, w, b) - y)**2 # loss = 1/2(h(x) - y)^2
    cost = (1/m)*np.sum(loss, axis=1, keepdims=True) # cost = mean of batch losses
    return cost
```

```python
def derivative(x, w, b, y):
    dw = ((1/m)*np.dot(x, (hypothesis(x, w, b) - y).T))[0][0] # d(cost)/dw =  (h(x) - y)(h(x))' = (h(x) - y)*x
    db = (1/m)*np.sum(hypothesis(x, w, b) - y) # d(cost)/db = (h(x) - y)(h(x))' = (h(x) -y)
    return dw, db
```

```python
def update(x, w, b, y, alpha):
    w = w - alpha*(derivative(x, w, b, y)[0]) # w := w - alpha * dw
    b = b - alpha*(derivative(x, w, b, y)[1]) # b := b - alpha * db
    return w, b
```

# Step 3. Training model

```python
# initialization of weights and bias
w = 0
b = 0
```

```python
count = 1
fig = plt.figure(figsize=(10,30))
for i in range(1,201):
    print ('cost =',cost(x, w, b, y)[0][0])
    w, b = update(x, w, b, y, alpha=0.01)
    print ('w =',w)
    print ('b =',b)
    print ('\n')
    
    if i % 20 == 0:
        ax = fig.add_subplot(5,2,count)
        ax.scatter(x,y)
        line_x = np.linspace(-4,4,100)
        ax.plot(line_x, line_x * w + b, 'r')
        count += 1
```

```python
for i in range(500):
    print ('cost =',cost(x, w, b, y)[0][0])
    w, b = update(x, w, b, y, alpha=0.01)

print ('w =',w)
print ('b =',b)  
```

### 1-2) Linear Regression with scikit-learn

### 1-3) Polynomial Linear Regression

```python
m = 100 # 데이터 개수(sample, example)
```

```python
x = np.random.randn(1,m)
y = 3 * (x**2) + np.random.rand(1,m) * 5
```

```python
plt.scatter(x,y)
plt.show()
```

```python
reshaped_x = x.reshape((100,1))
reshaped_y = y.reshape((100,1))
model = LinearRegression()
model.fit(reshaped_x,reshaped_y)
w = model.coef_[0][0]
b = model.intercept_[0]
print ('w =', w)
print ('b =', b)
```

```python
plt.scatter(x,y)
line_x = np.linspace(-3,4,100)
plt.plot(line_x, line_x * w + b, 'r')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
reshaped_x = x.reshape((100,1))
reshaped_y = y.reshape((100,1))
poly_x = poly.fit_transform(reshaped_x)
```

```python
model = LinearRegression()
model.fit(poly_x,reshaped_y)
```

```python
coef_1 = model.coef_[0][0]
coef_2 = model.coef_[0][1]
coef_3 = model.coef_[0][2]
print ('coefficient of one :', coef_1)
print ('coefficient of x :', coef_2)
print ('coefficient of x^2 :', coef_3)
```

```python
b = model.intercept_[0]
b
```

```python
plt.scatter(x,y)
line_x = np.linspace(-3,4,100)
plt.plot(line_x, (1 * coef_1) + (line_x * coef_2) + ((line_x ** 2) * coef_3) + b, 'r')
plt.show()
```
